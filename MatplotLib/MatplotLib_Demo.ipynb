```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x = np.arange(0,100)
y = x * 2
z = x ** 2

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_xlabel('x - axis')
ax.set_ylabel('y - axis')

plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.2,0.2])
ax1.plot(x,y, color = 'red')
ax2.plot(x,y, color = 'green')
ax1.set_title('Outer Title')
ax1.set_xlabel('Outer x - axis')
ax1.set_ylabel('Outer y - axis')

ax2.set_title('Inner Title')
ax2.set_xlabel('Inner x - axis')
ax2.set_ylabel('Inner y - axis')

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,0.4,0.4])

ax1.plot(x,z, color = 'red')
ax2.plot(x,y, color = 'green')

ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.set_title('Zoom')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])

plt.show()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y, color = 'blue', ls='--', lw=3)
axes[1].plot(x,z, color = 'red', lw=3)

plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12,3))
axes[0].plot(x,y, color = 'blue', ls='-', lw=3)
axes[1].plot(x,z, color = 'red',ls='--', lw=3)

plt.tight_layout()
```
